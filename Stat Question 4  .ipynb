What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook


Ans:
    Correlation is a statistical measure that describes the extent to which two variables change together. In other words, it quantifies the relationship between two variables, indicating whether they move in the same direction (positive correlation), opposite directions (negative correlation), or have no relationship (zero correlation).

Correlation values range between -1 and 1:

A correlation of 1 indicates a perfect positive correlation, meaning that as one variable increases, the other variable also increases proportionally.
A correlation of -1 indicates a perfect negative correlation, meaning that as one variable increases, the other variable decreases proportionally.
A correlation of 0 indicates no linear relationship between the variables.

Example Dataset: Let's consider a dataset containing the number of hours studied and the corresponding exam scores for a group of students.

```python
import pandas as pd

# Example dataset
data = {
    'Hours Studied': [2, 3, 4, 5, 6, 7, 8, 9],
    'Exam Score': [65, 67, 70, 72, 75, 78, 82, 85]
}

df = pd.DataFrame(data)
print(df)
```

We can visualize the relationship between hours studied and exam scores using a scatter plot. A scatter plot will show how the two variables are related and whether there is a correlation between them.

```python
import matplotlib.pyplot as plt

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Hours Studied'], df['Exam Score'], color='blue')
plt.title('Exam Score vs Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

In this scatter plot, if there is a positive correlation between hours studied and exam scores, we would expect to see the points forming an upward trend. Similarly, a negative correlation would show a downward trend, and no correlation would result in points scattered randomly without any clear trend.
